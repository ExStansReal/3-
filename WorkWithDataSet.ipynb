## Простой подстановщик
Теперь, я найду переменные в которых есть пустые значения коммандой `df.isnull().mean() * 100` а после использую простой подстановщик, чтобы заменить пустоту используя `SimpleImputer`. 

```python
df.isnull().mean() * 100
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer()

df['Year'] = imputer.fit_transform(df['Year'].values.reshape(-1,1))[:,0]

df.isnull().mean() * 100
```

Переменная Publisher не является int,double и т.п. По-этому использую `Replace`

```python
df = df.replace({"Publisher": np.nan}, "Not geted it")
df.isnull().mean() * 100
```
